```python
import sys
import os
import ctypes

print("="*60)
print("系统信息")
print("="*60)
print(f"Python版本: {sys.version}")
print(f"系统平台: {sys.platform}")

# 检查关键DLL是否存在
dlls = ["msvcp140.dll", "vcruntime140.dll", "vcruntime140_1.dll", 
        "cudart64_12.dll", "cudnn_ops_infer64_8.dll"]

for dll in dlls:
    try:
        ctypes.WinDLL(dll)
        print(f"✓ {dll} 存在")
    except OSError:
        print(f"✗ {dll} 缺失")

print("\n尝试导入torch...")
try:
    import torch
    print(f"✓ PyTorch导入成功")
    print(f"  版本: {torch.__version__}")
    print(f"  CUDA可用: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"  显卡: {torch.cuda.get_device_name(0)}")
except Exception as e:
    print(f"✗ 导入失败: {e}")
    
    # 检查torch目录
    import site
    torch_path = None
    for s in site.getsitepackages():
        test_path = os.path.join(s, "torch", "__init__.py")
        if os.path.exists(test_path):
            torch_path = os.path.dirname(os.path.dirname(test_path))
            break
    
    if torch_path:
        print(f"\nTorch安装路径: {torch_path}")
        lib_path = os.path.join(torch_path, "torch", "lib")
        if os.path.exists(lib_path):
            print(f"Torch库路径: {lib_path}")
            dll_files = [f for f in os.listdir(lib_path) if f.endswith('.dll')]
            print(f"找到的DLL文件: {len(dll_files)}个")
            
            # 检查shm.dll
            shm_path = os.path.join(lib_path, "shm.dll")
            if os.path.exists(shm_path):
                print(f"✓ shm.dll 存在，大小: {os.path.getsize(shm_path)} 字节")
            else:
                print(f"✗ shm.dll 不存在")
                # 列出lib目录内容
                print("Lib目录内容前10个:")
                for f in os.listdir(lib_path)[:10]:
                    print(f"  {f}")
print("="*60)
```
