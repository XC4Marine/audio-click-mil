```python
import pandas as pd
import numpy as np

# 读取Excel文件
df = pd.read_excel(r'D:\Dataset\IndoPacific_Humpback_Song\AllTrain.xlsx', sheet_name=0)

# 按 'Ori_file_num(No.)' 分组，并计算 'Train_duration(ms)' 的总和
duration_sum_by_file = df.groupby('Ori_file_num(No.)')['Train_duration(ms)'].sum()

# 将结果转换为DataFrame
result_df = duration_sum_by_file.reset_index()
result_df.columns = ['Ori_file_num(No.)', 'Total_Train_Duration(ms)']

# 按文件编号排序
result_df = result_df.sort_values('Ori_file_num(No.)').reset_index(drop=True)

# 提供的分钟数组
minutes_list = [30, 30, 30, 30, 30, 30, 30, 30, 30, 29.1, 12.8, 30, 30, 17.4, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30]

# 检查文件数量是否与分钟数组长度一致
if len(result_df) != len(minutes_list):
    print(f"警告: 文件数量({len(result_df)})与分钟数组长度({len(minutes_list)})不匹配!")
    # 只计算两者中较短的部分
    min_length = min(len(result_df), len(minutes_list))
    result_df = result_df.head(min_length)
    minutes_list = minutes_list[:min_length]

# 计算call rate
# 1. 将毫秒转换为分钟 (1分钟 = 60,000毫秒)
result_df['Total_Train_Duration(min)'] = result_df['Total_Train_Duration(ms)'] / 60000

# 2. 除以对应的分钟数
result_df['Provided_Duration(min)'] = minutes_list
result_df['Call_Rate'] = result_df['Total_Train_Duration(min)'] / result_df['Provided_Duration(min)']

# 3. 格式化结果，保留适当的小数位数
result_df['Total_Train_Duration(min)'] = result_df['Total_Train_Duration(min)'].round(3)
result_df['Call_Rate'] = result_df['Call_Rate'].round(4)

# 显示结果
print("不同Ori_file_num对应的统计结果：")
print("=" * 80)
print(result_df.to_string(index=False))
print("\n" + "=" * 80)
print(f"统计文件总数: {len(result_df)}")
print(f"平均Call Rate: {result_df['Call_Rate'].mean():.4f}")
print(f"Call Rate 范围: {result_df['Call_Rate'].min():.4f} - {result_df['Call_Rate'].max():.4f}")

# 可选：保存结果到新Excel文件
# result_df.to_excel('Call_Rate_Analysis.xlsx', index=False)
print(f"\n结果已保存到: Call_Rate_Analysis.xlsx")
```
